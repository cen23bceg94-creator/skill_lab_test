```python
# ============================================================
# FINSECURE - LOAN DEFAULT PREDICTION (FULL PIPELINE + AUC + SUBGROUPS)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    roc_auc_score, roc_curve, RocCurveDisplay
)

# ------------------------------------------------------------
# SECTION 1: LOAD DATA & DEFINE TARGET (DEFAULT vs PAID BACK)
# ------------------------------------------------------------

# Load your dataset (make sure loan_data.csv is in the same folder)
data = pd.read_csv("/content/sample_data/loan_data.csv")

# loan_paid_back: 1 = paid, 0 = not paid
# Define default: 1 = default, 0 = paid back
data["default"] = 1 - data["loan_paid_back"]

print("Target distribution (default):")
print(data["default"].value_counts(normalize=True))
print()

# Optional: quick bar plot of target distribution
plt.figure()
data["default"].value_counts().plot(kind="bar")
plt.xticks([0, 1], ["Paid back (0)", "Default (1)"], rotation=0)
plt.title("Class Distribution: Paid Back vs Default")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# SECTION 2: FEATURE ENGINEERING & PREPROCESSING PIPELINE
# ------------------------------------------------------------

# Numerical and categorical feature lists
num_cols = [
    "annual_income",
    "debt_to_income_ratio",
    "credit_score",
    "loan_amount",
    "interest_rate"
]

cat_cols = [
    "gender",
    "marital_status",
    "education_level",
    "employment_status",
    "loan_purpose"
    # "grade", # Removed as these columns are not in the DataFrame
    # "subgrade" # Removed as these columns are not in the DataFrame
]

# Features and target
X = data[num_cols + cat_cols]
y = data["default"]

# Keep ids to later attach predictions (for the internal test set)
ids = data["id"]

# Train-test split (stratified to preserve class balance)
X_train, X_test, y_train, y_test, ids_train, ids_test = train_test_split(
    X, y, ids,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# Preprocessing for numeric features: impute missing values + scale
num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# Preprocessing for categorical features: impute + one-hot encode
cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

# Combine numeric and categorical pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, num_cols),
        ("cat", cat_pipeline, cat_cols),
    ]
)

# ------------------------------------------------------------
# SECTION 3: MODEL DEVELOPMENT, TUNING & ROC (AUC) CURVE
# ------------------------------------------------------------

# Full pipeline: preprocessing + Logistic Regression
model_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",  # helpful for imbalance
        solver="lbfgs"
    ))
])

# Hyperparameter grid for tuning (regularization strength C)
param_grid = {
    "model__C": [0.01, 0.1, 1.0, 10.0]
}

grid_search = GridSearchCV(
    estimator=model_pipeline,
    param_grid=param_grid,
    scoring="roc_auc",   # AUC as optimization metric
    cv=5,
    n_jobs=-1,
    verbose=1
)

print("\nFitting GridSearchCV (this may take a bit)...")
grid_search.fit(X_train, y_train)

print("\nBest parameters:", grid_search.best_params_)
print("Best CV AUC:", grid_search.best_score_)

# Best model after tuning
best_model = grid_search.best_estimator_

# Predict probabilities for default (class 1)
y_train_proba = best_model.predict_proba(X_train)[:, 1]
y_test_proba = best_model.predict_proba(X_test)[:, 1]

# Overall AUC
train_auc = roc_auc_score(y_train, y_train_proba)
test_auc = roc_auc_score(y_test, y_test_proba)

print("\nTrain AUC:", train_auc)
print("Test AUC:", test_auc)

# ---- Plot ROC (AUC curve) for Train and Test ----
plt.figure()
RocCurveDisplay.from_predictions(
    y_train, y_train_proba, name="Train ROC"
)
RocCurveDisplay.from_predictions(
    y_test, y_test_proba, name="Test ROC"
)
plt.plot([0, 1], [0, 1], linestyle="--", label="Random baseline")
plt.title("ROC Curve - Train vs Test")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.show()

# Optional: histogram of predicted probabilities by class
plt.figure()
plt.hist(
    y_test_proba[y_test == 0],
    bins=30,
    alpha=0.6,
    label="True: Paid back (0)"
)
plt.hist(
    y_test_proba[y_test == 1],
    bins=30,
    alpha=0.6,
    label="True: Default (1)"
)
plt.title("Distribution of Predicted Default Probabilities (Test Set)")
plt.xlabel("Predicted probability of default")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
plt.figure()
sns.histplot(x=y_test_proba, hue=y_test, bins=30)
plt.title("Dist. of Default Probabilities")
plt.xlabel("Predicted Probability")
plt.show()

# ---- Choose a threshold (optional, for business use) ----
# AUC doesn't need a threshold, but we can choose one (Youden's J) for classification
fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)
youden_index = tpr - fpr
best_idx = np.argmax(youden_index)
best_threshold = thresholds[best_idx]

print("\nBest threshold by Youden's J:", best_threshold)
y_test_pred = (y_test_proba >= best_threshold).astype(int)

# ------------------------------------------------------------
# SECTION 4: SUBGROUP ANALYSIS (AUC BY EDUCATION & PURPOSE)
# ------------------------------------------------------------

# Attach predictions and true labels to the test feature set
X_test_eval = X_test.copy()
X_test_eval["default"] = y_test.values
X_test_eval["pred_proba"] = y_test_proba

# ---------- AUC by education_level ----------
edu_auc = {}

for level, group in X_test_eval.groupby("education_level"):
    # Need both classes (0 and 1) to compute AUC meaningfully
    if group["default"].nunique() < 2:
        edu_auc[level] = np.nan
    else:
        auc_val = roc_auc_score(group["default"], group["pred_proba"])
        edu_auc[level] = auc_val

edu_auc_series = pd.Series(edu_auc).sort_values(ascending=False)

print("\nAUC by education_level:")
print(edu_auc_series)

# Plot AUC by education level as bar chart
plt.figure()
edu_auc_series.plot(kind="bar")
plt.title("AUC by Education Level (Test Set)")
plt.xlabel("Education Level")
plt.ylabel("AUC")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# ---------- AUC by loan_purpose ----------
purpose_auc = {}

for purpose, group in X_test_eval.groupby("loan_purpose"):
    # Require at least 2 classes and a minimum sample size
    if group["default"].nunique() < 2 or len(group) < 50:
        continue
    auc_val = roc_auc_score(group["default"], group["pred_proba"])
    purpose_auc[purpose] = auc_val

if len(purpose_auc) == 0:
    print("\nNot enough data to compute AUC by loan_purpose (after filtering).")
    purpose_auc_series = pd.Series(dtype=float)
else:
    purpose_auc_series = pd.Series(purpose_auc).sort_values(ascending=False)

    print("\nAUC by loan_purpose (filtered groups):")
    print(purpose_auc_series)

    # Top 3 and bottom 3
    top_3 = purpose_auc_series.head(3)
    bottom_3 = purpose_auc_series.tail(3)

    print("\nTop 3 loan_purpose by AUC:")
    print(top_3)

    print("\nBottom 3 loan_purpose by AUC:")
    print(bottom_3)

    # Plot AUC by loan purpose
    plt.figure()
    purpose_auc_series.plot(kind="bar")
    plt.title("AUC by Loan Purpose (Test Set, Filtered)")
    plt.xlabel("Loan Purpose")
    plt.ylabel("AUC")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# SAVE PER-ID PREDICTIONS (OPTIONAL)
# ------------------------------------------------------------

predictions_df = pd.DataFrame({
    "id": ids_test.values,
    "default_probability": y_test_proba,
    "loan_paid_back_probability": 1 - y_test_proba
})

predictions_df.to_csv("loan_default_test_predictions.csv", index=False)
print("\nSaved per-id probabilities to 'loan_default_test_predictions.csv'.")

print("\n=== SCRIPT FINISHED SUCCESSFULLY ===")
```
